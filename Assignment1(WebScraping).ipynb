# Question 1

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

pd.set_option('display.max_colwidth', None)  #to display the entire url

def getData():
    url = 'https://timesofindia.indiatimes.com/home/headlines'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    headlineElements = soup.find_all('span', attrs={'class': "w_tle"})
    
    df = pd.DataFrame(columns=['Headline', 'Link'])  
    
    data = []  
    
    for h in headlineElements[:10]:
        headline = h.text if h and h.text else ' '  
        link = h.find('a')['href'] if h.find('a') else ' '
        full_link = 'https://timesofindia.indiatimes.com/india' + link if link.startswith('/') else link
        
        data.append({'Headline': headline, 'Link': full_link})  
    
    df = pd.concat([df, pd.DataFrame(data)], ignore_index=True)  
    
    return df


df=getData()
print(df)


```

# Question 2

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

pd.set_option('display.max_colwidth', None)
df = pd.DataFrame(columns=['jobTitle', 'jobCompany', 'jobLink'])

def get_data(location, role):
    url = "https://www.linkedin.com/jobs/search?keywords="+str(role)+"&location="+str(location)+"&geoId=&trk=public_jobs_jobs-search-bar_search-submit&original_referer=https%3A%2F%2Fwww.linkedin.com%2Fjobs%2Fsearch%3Ftrk%3Dguest_homepage-basic_guest_nav_menu_jobs%26position%3D1%26pageNum%3D0&position=1&pageNum=0"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    for d in soup.find_all('div', attrs={'class': 'base-search-card'}):
        job_title = d.find('h3', attrs={'class':'base-search-card__title'})
        job_company = d.find('h4', attrs={'class':'base-search-card__subtitle'})
        job_link = d.find('a', attrs={'class':'base-card__full-link absolute top-0 right-0 bottom-0 left-0 p-0 z-[2]'})
        job_data = []
        if job_title is not None:
            job_data.append(job_title.text.strip())
        else:
            job_data.append(' ')
        if job_company is not None:
            job_data.append(job_company.text.strip())
        else:
            job_data.append(' ')
        if job_link is not None:
            job_data.append('https://www.linkedin.com' + job_link.get('href'))
        else:
            job_data.append(' ')
        df.loc[len(df)] = job_data
    return df

result_df = get_data("india", "frontend developer")
print(result_df)
print(result_df.shape)
```
